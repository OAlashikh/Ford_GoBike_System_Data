## Univariate Exploration

> In this particular part we are investigating the distributions of individual variables.

```python
binsize = 500
bins = np.arange(0, df['duration_sec'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df, x = 'duration_sec', bins = bins)
plt.title('Distribution of Trip Durations')
plt.xlabel('Duration_sec')
plt.ylabel('Number of Trips')
plt.axis([-500, 10000, 0, 90000])
plt.show()
```

##### there's a long tail in the distribution, so let's put it on a log scale instead

```python

log_binsize = 0.05
bins_log = 10 ** np.arange(2.4, np.log10(df['duration_sec'].max()) + log_binsize, log_binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df, x = 'duration_sec', bins = bins_log)
plt.title('Distribution of Trip Durations')
plt.xlabel('Duration_sec')
plt.ylabel('Number of Trips')
plt.xscale('log')
plt.xticks([500, 1e3, 2e3, 5e3, 1e4], [500, '1k', '2k', '5k', '10k'])
plt.axis([0, 10000, 0, 15000])
plt.show()
```

Trip duration is mostly centered on values which are less than 2000 seconds with peak around 600 seconds. Number of Trips values first increases starting from around 8000 values to 12500 values at around 600 seconds but then starts to fall below at 2000 values under 2000 seconds.

I'll now move on to the other variables in the dataset to look into the duration dependency on start station and end station.

```python
sorted(df.start_station_id.unique())
```

```python
S = []

all_start_station_ids = sorted(df.start_station_id.unique())
for x in all_start_station_ids :
    S.append(df[df.start_station_id == x].duration_sec.sum()) 
total_duration = pd.Series(S)

plt.figure(figsize = [20, 8])
sb.lineplot(x = df['start_station_id'], y = total_duration)
plt.xticks(range(0, 401, 10))
plt.title('Trip Duration and Start Station')
plt.xlabel('Start Station')
plt.ylabel('Total Duration')
plt.show()
```

```python
E = []

all_end_station_ids = sorted(df.end_station_id.unique())
for x in all_end_station_ids :
    E.append(df[df.end_station_id == x].duration_sec.sum()) 
total_duration = pd.Series(E)

plt.figure(figsize = [20, 8])
sb.lineplot(x = df['end_station_id'], y = total_duration)
plt.xticks(range(0, 401, 10))
plt.title('Trip Duration and End Station')
plt.xlabel('End Station')
plt.ylabel('Total Duration')
plt.show()
```

```python
fig = plt.figure(figsize = [8,6]);
ax = sb.pointplot(data = df, x = 'start_station_latitude', y = 'duration_min', hue = 'user_type',
           palette = 'Blues', linestyles = '');
plt.title('Biking durations with the start station latitude')
plt.xlabel('start station latitude')
plt.ylabel('Duration in minutes')
plt.yscale('log')
plt.yticks([1, 2, 5, 10, 20, 50, 100], [1, 2, 5, 10, 20, 50, 100]);
plt.xticks([0,10,20,35,50,65,80,95,110,125,140,155,170,185,200,215,230,245,260,275,290],[0,10,20,35,50,65,80,95,110,125,140,155,170,185,200,215,230,245,260,275,290]);
ax.set_yticklabels([],minor = True);
```

```python
x_bin = np.arange( df['member_birth_year'].min(), df['member_birth_year'].max()+5, 5)
y_bin = np.arange(0, (838 + 1.5*474)+500, 500)
_temp = df.sample(35000, random_state=42)
m = sb.FacetGrid(data = _temp, col = 'user_type',height=5.4)
m.map(plt.hist2d, 'member_birth_year','duration_sec',cmin=0.5,cmap = 'viridis_r',bins=[x_bin, y_bin]);
plt.ylim((0,(838 + 1.5*474)));
plt.colorbar();
plt.xlabel('Year of Birth');
plt.ylabel('Duration in Seconds');
```

- Customers usually takes up to from 500 to 1000 Seconds in thier Rides
- and they are all between 1980 and 1995 Years old

```python
cat_means = df.groupby(['member_gender', 'user_type']).mean()['duration_sec']
cat_means = cat_means.reset_index(name = 'Duration_avg')
cat_means = cat_means.pivot(index = 'user_type', columns = 'member_gender',
                            values = 'Duration_avg')
sb.heatmap(cat_means, annot = True, fmt = '.3f',
           cbar_kws = {'label' : 'mean(Duration_avg)'});
```

- As you can see, the average Duration is highest for customers that identify as "other" ( top Right )
- where the average duration is lowest for male subscribers ( bottom middle )
